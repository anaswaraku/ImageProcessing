Pgm to open an image, display it and retrive its dimensions using opecv

```python
import cv2
```

```python
img = cv2.imread("myimage.jpg", cv2.IMREAD_COLOR)
cv2.imshow("image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)# BGR (Blue-Green-Red) to RGB (Red-Green-Blue) format
plt.imshow(img_rgb)
plt.axis('off')  # Turn off the axis
plt.show()
```

```python
print("Image Dimension:",img.shape)
```

```python
gray_img= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("image",gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
